# Extract frame by steps

```python
baseDir = 'member-vids'
# videos = ['duy','long','tung','vinh']
videos = ['vinh']
print(videos)
for vidName in videos:
    vid = '%s/%s.mp4'%(baseDir,vidName)
    c_frame = 0
    offset=300  # prevent conflict
    step=5  #capture image every `step` frames
    saveDir = 'output'
    # Create a VideoCapture object and read from input file
    # If the input is the camera, pass 0 instead of the video file name
    cap = cv2.VideoCapture(vid)
    # Check if camera opened successfully
    if (cap.isOpened()== False): 
        print("Error opening video stream or file")
    ret, frame = cap.read()
    if ret == True:
        # Display the resulting frame
        frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        frame = cv2.resize(frame,(0,0),fx=.5,fy=.5) # resizes image to half
        cv2.imshow(vidName,frame)
        cv2.imwrite('%s/%s%08d.jpg'%(saveDir,vidName,offset+c_frame),frame)

    print('Press "n" to go to next frame')

    # Read until video is completed
    while(cap.isOpened()):
        # if cv2.waitKey(25) & 0xFF == ord('s'):
        #     cv2.imwrite('%s%08d.jpg'%(vidName,c_frame),frame)
        # if cv2.waitKey(25) & 0xFF == ord('q'):
        #     break
        # if cv2.waitKey(25) & 0xFF == ord('n'):  
            # Capture frame-by-frame
        ret, frame = cap.read()
        if ret == True:
            c_frame = c_frame+1
            # Display the resulting frame
            if c_frame%5 == 0:

                frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
                frame = cv2.resize(frame,(0,0),fx=.5,fy=.5) # resize to half
                cv2.imwrite('%s/%s%08d.jpg'%(saveDir,vidName,offset+c_frame),frame)
        # Break the loop
        else: 
            break
    # When everything done, release the video capture object
    cap.release()

    # Closes all the frames
    cv2.destroyAllWindows()
```

## Extract frames manually

```python
baseDir = 'member-vids/new-vids'
videos = ['tung%d'%i for i in range(2,7)];
print(videos)
for i in range(len(videos)):
    vid = '%s/%s.mp4'%(baseDir,videos[i])
    print(vid)
    c_frame = 0
    offset=i  # prevent conflict
    saveDir = 'output'
    # Create a VideoCapture object and read from input file
    # If the input is the camera, pass 0 instead of the video file name
    cap = cv2.VideoCapture(vid)
    # Check if camera opened successfully
    if (cap.isOpened()== False): 
        print("Error opening video stream or file")
    ret, frame = cap.read()
    if ret == True:
        frame = cv2.resize(frame,(0,0),fx=.5,fy=.5)
        cv2.imshow(videos[i],frame)
    print('Press "n" to go to next frame')
    # Read until video is completed
    while(cap.isOpened()):
        if cv2.waitKey(25) & 0xFF == ord('s'):
            cv2.imwrite('%s/%s%01d%08d.jpg'%(saveDir,videos[i],offset,c_frame),frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
        if cv2.waitKey(25) & 0xFF == ord('n'): 
            print('pressed "n"')
            ret, frame = cap.read()
            if ret == False:
                print('Error getting video frame.')
                break
            c_frame = c_frame+1
            frame = cv2.resize(frame,(0,0),fx=.5,fy=.5)
            cv2.imshow(videos[i],frame)
    # When everything done, release the video capture object
    cap.release()

    # Closes all the frames
    cv2.destroyAllWindows()
```
